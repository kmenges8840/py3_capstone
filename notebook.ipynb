```python
import pandas as pd
```

```python
raw = pd.read_excel('data/Nucor.xlsx',header=None)
raw 
```

```python
def clean_format(dataframe):
    dataframe = dataframe.dropna(thresh=1)
    dataframe = dataframe.dropna(axis='columns',how='all')
    dataframe.fillna("",inplace=True)
    dataframe = dataframe.reset_index(drop=True)
    dataframe['COMBINED'] = dataframe[dataframe.columns[0:]].apply(
    lambda x: ' '.join(x.astype(str)),
    axis=1)
    return dataframe['COMBINED']
```

```python
test = clean_format(raw)
test
```
